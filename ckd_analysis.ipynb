```python
import pandas as pd
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt2
%matplotlib inline
```

```python
test_data = pd.read_csv('./data/testing.csv')

test_data_sort = test_data.sort_values(by='DSMED53')
test_data_clean = test_data_sort[:-12]

meds_false = test_data_clean[:1248].sort_values(by='Kidney_Disorder')
meds_true = test_data_clean[1249:].sort_values(by='Kidney_Disorder')

meds_false_kd_false = meds_false[:1141]
meds_false_kd_true = meds_false[1142:]

meds_true_kd_false = meds_true[:212]
meds_true_kd_true = meds_true[213:]
```

```python
# test_data_clean.shape 1503
# meds_false.shape 1248 + meds_true.shape 254 = 1502
# (meds_false_kd_false.shape 1141 + meds_false_kd_true.shape 106) 1247
# (meds_true_kd_false.shape 212 + meds_true_kd_true.shape 41) 253
# 1247 + 253 = 1500
```

```python
# Disregard
kidney_transform_false = meds_false['Kidney_Disorder'].replace(1, 2)
kidney_transform_false_finish = kidney_transform_false.replace(0, 1)

kidney_transform_true = meds_true['Kidney_Disorder'].replace(1, 2)
kidney_transform_true_finish = kidney_transform_true.replace(0, 1)
```

## SEX

```python
cm = confusion_matrix(meds_false['SEX'], kidney_transform_false_finish)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.show()
```
